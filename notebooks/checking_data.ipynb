```python
import pandas as pd
```

```python
df = pd.read_csv("../data/Titanic-Dataset.csv")
df.head()
```

```python
print(f"Dimensions, (Rows, cols): {df.shape}")
print(f"Column names: {df.columns}")

# Lowering the column names:
df.columns = df.columns.str.lower()

print(f"\nColumn names after lowering: {df.columns}")
```

```python
print("Datatypes:\n\n",df.dtypes)
```

```python
# Checking the unique values;
for col in df.columns:
    print(f"Column: {col}")
    print(f"Unique values count: {len(df[col].unique())}")

    if len(df[col].unique()) <= 10:
        print(f"Unique values: {df[col].unique()}")
    print()
```

```python
# Handling the null values:
print(df.isnull().sum())
```

```python
# handling age:

print(df["age"].min())
print(df["age"].max())

# Filling the age missing data with Median:
df["age"] = df["age"].fillna(df["age"].median())
```

```python
# Using Most frequent to handle embarked
df["embarked"] = df["embarked"].fillna(df["embarked"].mode()[0])
```

```python
# Dropping the column table, as most of the data is missing
df = df.drop(columns=["cabin"])
```

```python
print("Checking the Null values again:\n\n",df.isnull().sum())
```

```python
# Saving the CLEANED CSV FILE

df.to_csv("../data/titanic_cleaned.csv", index=False)
```
